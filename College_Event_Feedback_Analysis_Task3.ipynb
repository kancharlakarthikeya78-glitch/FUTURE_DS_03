## 1. Import Libraries

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from textblob import TextBlob
sns.set()
```

## 2. Load Dataset

```
df = pd.read_csv('college_event_feedback.csv')
df.head()
```

## 3. Rating Analysis

```
df.groupby('Event_Name')['Rating'].mean().plot(kind='bar')
plt.title('Average Rating per Event')
plt.show()
```

## 4. Sentiment Analysis

```
def get_sentiment(text):
    return TextBlob(text).sentiment.polarity

df['Sentiment_Score'] = df['Feedback_Comment'].apply(get_sentiment)
df['Sentiment'] = df['Sentiment_Score'].apply(lambda x: 'Positive' if x > 0 else 'Negative' if x < 0 else 'Neutral')
df.head()
```

## 5. Sentiment Distribution

```
df['Sentiment'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Sentiment Distribution')
plt.ylabel('')
plt.show()
```

## 6. Insights & Recommendations
- Workshops received high satisfaction
- Cultural events were most liked
- Seminars need better engagement
